```python
from chase.base import CHASEModel, CHASEAlternateStoppingModel
```

```python
options = np.array([[[  8.        ,   .9],
                     [ 28.        ,   0.1]],
                    [[  8.  ,   0.9],
                     [  78.  ,   0.1]]])
```

### Basic optional stopping model

```python
# specify the model, including the 
# models used for the drift and the 
# starting distribution
m = CHASEModel(drift='ev',
               startdist='indifferent')

# minimal parameter set
pars = {'theta': 5,
        'p_stay': .3}

# run the model
result = m(options, pars)
```

```python
fig, ax = plt.subplots(figsize=(6,4))
ax.plot(result['p_stop_cond'][:,0], label='p(samplesize|L)')
ax.plot(result['p_stop_cond'][:,1], label='p(samplesize|H)')
ax.set_xlabel('Sample size')
ax.legend()
plt.show()
```

### Optional stopping model with CPT drift rate

```python
m = CHASEModel(drift='cpt',
               startdist='indifferent')

# minimal parameter set
pars = {'theta': 5,
        'p_stay': .3,
        'pow_gain': 0.8}

# run the model
result = m(options, pars)
```

```python
fig, ax = plt.subplots(figsize=(6,4))
ax.plot(result['p_stop_cond'][:,0], label='p(samplesize|L)')
ax.plot(result['p_stop_cond'][:,1], label='p(samplesize|H)')
ax.set_xlabel('Sample size')
ax.legend()
plt.show()
```

### Truncated normal sample size distribution, with CPT drift rate

```python
m = CHASEAlternateStoppingModel(drift='cpt',
                                startdist='indifferent',
                                stoprule='truncatednormal')

# minimal parameter set
pars = {'theta': 30,
        'p_stay': .3,
        'mu': 20,
        'sc': 5,
        'pow_gain': 0.8}

# run the model
result = m(options, pars)
p_stop = m.stoprule.dist(pars)
```

```python
plt.matshow(result['states_t'].transpose())
```

```python
fig, ax = plt.subplots(figsize=(6,4))
ax.plot(p_stop, label='p(samplesize)', color='black')
ax.set_xlabel('Sample size')
ax.legend()
ax.set_xlim(0, 100)
plt.show()
```
